```python
FULL = 5
stack = [0,1,2,3,4]
top = 4

def isStackFull():
  global FULL,stack,top
  if (top >= FULL):
    return True
  else:
    return False

def isStackEmpty():
  global FULL,stack,top
  if (top<=-1):
    return True
  else:
    return False

def push(a):
  global FULL,stack,top
  if(isStackFull()):
    print("Stack is Full")
    return
  top += 1
  stack[top] = a

def pop():
  global FULL,stack,top
  if(isStackEmpty()):
    print("Stack is empty")
    return
  a = stack[top]
  stack[top] = None
  top -= 1
  return a

print(pop())
```
