```python
class Node1():
  def __init__(self):
    self.data = None
    self.link = None

class Node2():
  def __init__(self):
    self.data = None
    self.link = None

n0 = Node1()
n0.data = 3

n1 = Node1()
n1.data = 5

n2 = Node1()
n2.data = 10

n3 = Node1()
n3.data = 21

k0 = Node2()
k0.data = 4

k1 = Node2()
k1.data = 6

k2 = Node2()
k2.data = 15

k3 = Node2()
k3.data = 26

k4 = Node2()
k4.data = 33

n0.link = n1
n1.link = n2
n2.link = n3

k0.link = k1
k1.link = k2
k2.link = k3
k3.link = k4

head1 = n0
head2 = k0

def Merge(a, b):
    if a == None:
        return b
    if b == None:
        return a
    if a.data < b.data:
        result = a
        result.link = Merge(a.link, b)
    else:
        result = b
        result.link = Merge(a, b.link)
    return result

current = Merge(head1,head2)
while current.link != None:
  print(current.data)
  current = current.link
print(current.data)
```
